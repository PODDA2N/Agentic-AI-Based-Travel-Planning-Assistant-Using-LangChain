# **Project Summary -**

The Agentic AI Based Travel Planning Assistant is an intelligent AI-powered travel recommendation and itinerary generation system developed using Python, FastAPI, LangChain, OpenAI, and real-time APIs. The project is designed to automate the complete travel planning process by integrating flight search, hotel recommendation, tourist attraction discovery, weather forecasting, and budget estimation into a single AI-driven platform.

The system uses an Agentic AI architecture, where an AI agent autonomously decides which tools and APIs to use based on user travel queries. The assistant analyzes user preferences such as source location, destination, travel duration, and budget, then generates a structured travel plan including selected flights, hotel recommendations, weather forecasts, day-wise itineraries, and estimated trip costs.

#### Budget Estimation Module

```python
total_budget = estimate_budget(
        selected_flight,
        selected_hotel,
        days
    )
```

#### What is the purpose of the Budget Estimation Module?

The purpose of the Budget Estimation Module in the Agentic AI-Based Travel Planning Assistant project is to calculate the estimated total cost of the trip based on travel expenses such as flight cost, hotel accommodation, and daily local expenses.

The Budget Estimation Module helps users understand the approximate travel budget before finalizing their trip plans. It enables the AI system to generate affordable, optimized, and realistic travel recommendations according to the user’s financial preferences.

The module calculates expenses including:

flight ticket cost
hotel cost per night
local transportation expenses
food and daily travel expenses

The Budget Estimation Tool was created to automate cost calculation and simplify financial planning for travelers.

In the project workflow, the LangChain AI agent dynamically calls the Budget Estimation Tool after retrieving:

flight information
hotel recommendations
travel duration details

The module then calculates the overall trip budget using the formula:

Total Budget=Flight Cost+(Hotel Price Per Night×Days)+(1000×Days)

Where:

Flight Cost = selected flight ticket price
Hotel Price Per Night × Days = accommodation expenses
1000 × Days = estimated local food and transportation expenses

The Budget Estimation Module helps:

improve travel planning efficiency
provide budget-friendly recommendations
compare travel options
optimize trip expenses
support decision-making
